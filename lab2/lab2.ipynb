```python
import pandas as pd
import numpy as np
from apriori import *
```

```python
path = "dataset/T10I4D100K.dat"
df = pd.read_csv(path, header=None, names=["basket"])
# convert each basket to a list of items
df["basket"] = df["basket"].apply(lambda x: list(map(int, x.split())))

df.head()
```

```python
s = 1000
apriori_instance = apriori(df["basket"], s)
all_frequent_itemsets = apriori_instance.get_frequent_itemsets()
```

```python
for i, itemset in enumerate(all_frequent_itemsets):
    print(f"Found {len(itemset)} itemsets of length {i+1}")
    
```

```python
for itemsets_of_size_k in all_frequent_itemsets:
    for itemset, count in itemsets_of_size_k.items():
        print(f"Itemset: {itemset}, Count: {count}")
```
